```python
import pandas as pd
```

```python
tweets = (
    pd
    .read_csv('gdefournas.csv')
    .set_index('id')
)

tweets
```

```python
sentiments = (
    pd
    .read_csv('gdefournasxlm-robertaok_withnature.csv')
    .set_index('id')
)

sentiments
```

```python
(
    sentiments
    .assign(
        local_time = lambda df: pd.to_datetime(df.local_time),
        date = lambda df: pd.to_datetime(df.local_time).dt.to_period('W')
    )
    .reset_index()
    .groupby(['date', 'sentiment'])
    [['id']]
    .count()
    .reset_index()
    .pivot_table(
        index='date',
        columns='sentiment'
    )
    .fillna(0)
    .rolling(50)
    .mean()
    .plot
    .area(
        stacked=True,
        figsize=(15,10)
    )
)
```

```python
(
    sentiments
    .assign(
        local_time = lambda df: pd.to_datetime(df.local_time),
        date = lambda df: pd.to_datetime(df.local_time).dt.to_period('W')
    )
    .reset_index()
    .groupby(['date', 'sentiment'])
    [['retweet_count']]
    .sum()
    .reset_index()
    .pivot_table(
        index='date',
        columns='sentiment'
    )
    .fillna(0)
    .rolling(50)
    .mean()
    .plot
    .area(
        stacked=True,
        figsize=(15,10)
    )
)
```
